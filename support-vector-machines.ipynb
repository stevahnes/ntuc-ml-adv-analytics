# Support Vector Machines

Objective:

- Dataset which has data pertaining to certain images. These images are digits between 0-9 represented as columns of data.

- Given a representation of a digit as columns of data, train a model to predict the right digit.

- Image is originally represented as a matrix (rows * columns).

    - See what an image looks like.

    - Use the flattened version of the image to train.

    - For example: if we have an 8x8 image, it means that when flattened it will contain 64 columns.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm, datasets, metrics
```

```python
# Load the dataset, where every digit is represented as 64 columns (known from the data)
digits = datasets.load_digits() # dictionary datatype
```

```python
# Look at a few samples of the data
num_data_sample = 10
for i in range(0,num_data_sample):
    # Create some plot space for the five samples
    plt.subplot(2,np.ceil(num_data_sample/2),i+1)
    plt.axis('off')
    # Translate 64 columns to 8x8 representation for visualization
    img_side = 8
    img = np.reshape(digits.data[i], (img_side, img_side))
    # Plot the images
    plt.imshow(img)

plt.show()
```

```python
# Perform train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size = 0.2)
```

```python
# Train the SVM model
svm_classifier = svm.SVC() # Create an instance
svm_classifier.fit(X_train, y_train)
```

```python
# Test the model
y_pred = svm_classifier.predict(X_test)
```

```python
# Evaluate the model

# Accuracy
test_acc = metrics.accuracy_score(y_test, y_pred)
print("Testing Accuracy: ", test_acc)
```

```python
# Confusion Matrix
conf_mat = metrics.confusion_matrix(y_pred, y_test)
#Plot the matrix
sns.heatmap(conf_mat.T, annot=True, cbar=False)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
