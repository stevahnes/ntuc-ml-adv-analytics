# Decision Trees

Objective:

- Predict the type of flower from the IRIS dataset

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn import datasets, metrics
```

```python
# Load IRIS dataset
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names) # Inputs
iris_target = iris.target # Output
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_df, iris_target, test_size=0.2, shuffle=True)
```

```python
# Train the model
from sklearn.tree import DecisionTreeClassifier
decision_tree_classifier = DecisionTreeClassifier() # Create an instance of the model
decision_tree_classifier.fit(X_train, y_train)
```

```python
# Visualize the model
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Allocate plot area for Decision Tree
plt.figure(figsize=(25,10))
decision_tree_plot = plot_tree(decision_tree_classifier, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, fontsize=14)
plt.show()
```

```python
# Test the model
y_pred = decision_tree_classifier.predict(X_test)
```

```python
test_acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", test_acc)
```

```python
import seaborn as sns
# Confusion Matrix
conf_mat = metrics.confusion_matrix(y_test, y_pred)
#Plot the matrix
sns.heatmap(conf_mat.T, annot=True, cbar=False, xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
