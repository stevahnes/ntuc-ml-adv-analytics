# Decision Trees

Objective:

- Predict the type of flower from the IRIS dataset

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# Load IRIS dataset
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names) # Inputs
iris_target = iris.target # Output
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_df, iris_target, test_size=0.2, shuffle=False)
```

```python
# Train the model
from sklearn.tree import DecisionTreeClassifier
decision_tree_classifier = DecisionTreeClassifier() # Create an instance of the model
decision_tree_classifier.fit(X_train, y_train)
```

```python
# Visualize the model
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Allocate plot area for Decision Tree
plt.figure(figsize=(25,10))
decision_tree_plot = plot_tree(decision_tree_classifier, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, fontsize=14)
plt.show()
```
