# Data Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

# Logistic Regression

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test)
```

```python
test_acc_log_reg = metrics.accuracy_score(y_test, y_pred_log_reg)
print("Accuracy: ", test_acc_log_reg)
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred_log_reg)
sns.heatmap(conf_mat.T, annot=True, cbar=False, xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```

# SVM

```python
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
```

```python
y_pred_svm = svm_classifier.predict(X_test)
```

```python
test_acc_svm = metrics.accuracy_score(y_test, y_pred_svm)
print("Testing Accuracy: ", test_acc_svm)
```

```python
conf_mat = metrics.confusion_matrix(y_test, y_pred_svm)
sns.heatmap(conf_mat.T, annot=True, cbar=False, xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
